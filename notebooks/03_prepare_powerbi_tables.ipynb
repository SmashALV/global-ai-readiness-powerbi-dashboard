```python
import pandas as pd
#Cargar Dimensión pais
dim_country = pd.read_csv("../data/processed/dim_country.csv")
ai_scores = pd.read_csv("../data/processed/ai_readiness_scores.csv")
```

```python
#Crear una columna Inicial
latam_iso3 = [
    "ARG", "BOL", "BRA", "CHL", "COL", "CRI", "CUB", "DOM", "ECU",
    "SLV", "GTM", "HND", "MEX", "NIC", "PAN", "PRY", "PER", "URY",
    "VEN", "BLZ", "HTI", "JAM", "TTO", "BRB", "BHS", "GUY", "SUR"
]

dim_country["Region"] = dim_country["ISO3"].apply(
    lambda x: "Latin America & Caribbean" if x in latam_iso3 else "Other Regions"
)

dim_country["IsLatAm"] = dim_country["ISO3"].isin(latam_iso3)
dim_country["IsPeru"] = dim_country["ISO3"].eq("PER")
```

```python
#Guardar la nueva versión
dim_country.to_csv("../data/processed/dim_country_enriched.csv", index=False)
```

```python
#Agregar Countrykey al score
ai_scores_model = ai_scores.merge(
    dim_country[["CountryKey", "Country", "ISO3", "Region", "IsLatAm", "IsPeru"]],
    on=["Country", "ISO3"],
    how="left"
)

ai_scores_model.to_csv("../data/processed/ai_readiness_scores_model.csv", index=False)
```
